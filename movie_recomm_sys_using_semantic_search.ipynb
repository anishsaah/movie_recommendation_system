```python
import pandas as pd
df = pd.read_csv("/content/movies.csv")
df.dropna(inplace=True)
```

```python
df.head(2)
```

```python
feature = df.title + " " + df['genres'].apply(lambda x: x.replace("|", " "))
feature = feature.str.lower()
feature
```

```python
pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(feature.values).tolist()
```

```python
df.dtypes
```

```python
ids = df.movieId.astype(str).values.tolist()
metadata = df.to_dict('records')
```

```python
import chromadb
client = chromadb.Client()
```

```python
collection = client.create_collection("vector_db")
```

```python
collection.add(
    ids = ids,
    documents = feature.values.tolist(),
    embeddings = embeddings,
    metadatas=metadata
    )
```

```python
user_feature = input("What kind of genres of movies do you like?")
```

```python
user_feature
```

```python
#Semantic Search
recommendations = collection.query(query_texts= [user_feature], n_results=3)
```
